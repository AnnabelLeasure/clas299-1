```scala
val url = "http://shot.holycross.edu/ada-vases.tsv"
```

```scala
import scala.io.Source
val lines = Source.fromURL(url).getLines.toVector
lines.size
```

```scala
val byColumns = lines.map(ln => ln.split("\t").toVector)
byColumns.size
```

```scala
val trimmed = byColumns.filter(row => row.size == 23).map( v => Vector(v(1), v(5), v(6), v(7), v(21), v(22)))
```

```scala
// how many full records?
trimmed.size
trimmed.head
```

```scala
val shapes = trimmed.map(columns => columns(1))
```

```scala
val shapesGrouped = shapes.groupBy(shape => shape)
```

```scala
val shapesCounts = shapesGrouped.map{ case (d, v) => (d, v.size) }
```

```scala
val shapesHisto = shapesCounts.toVector.sortBy(frequency => frequency._2).reverse
```

```scala
val unique = shapes.distinct
```

```scala
println(unique)
```

```scala
// extracting amphora entries
val amphora = trimmed.filter(vase => vase(1).toLowerCase.contains("amphora"))
```

```scala
// removing panathenaic amphora entries
val amphoraReal = amphora.filterNot(vase => vase(1).toLowerCase.contains("panathenaic"))
```

```scala
amphora.size
amphoraReal.size
```

```scala
val amphoraCut = trimmed.filter(shp => shp(1) == "AMPHORA")
```

```scala
val amphoraCut2 = amphoraReal.filter(vase => vase(1).toLowerCase.contains("?"))
```

```scala
val amphoraCut3 = amphoraCut2.filterNot(vase => vase(1).toLowerCase.contains("neck"))
```

```scala
// how many amphora entries with unspecified shape?
amphoraCut.size
amphoraCut3.size
```

```scala
// how many neck amphoras?
val amphoraNeck = amphoraReal.filter(vase => vase(1).toLowerCase.contains("neck"))
amphoraNeck.size
```

```scala
// how many bail amphoras?
val amphoraBail = amphoraReal.filter(vase => vase(1).toLowerCase.contains("bail"))
amphoraBail.size
```

```scala
val csv = amphoraReal.map ( v => v.mkString(";")).mkString("\n")
```

```scala
val header = "number;shape;provenance;date;latitude;longitude\n"
```

```scala
println(header + csv)
```
